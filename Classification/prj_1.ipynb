```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport

df = pd.read_csv("data.csv")
profile = ProfileReport(df, title="My Data Report")

# Save as HTML file to view in browser
profile.to_file("data_report.html")
print("Report saved as data_report.html - open it in your browser")
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()


```

```python
# Check column names and shape
print("Columns in dataframe:")
print(df.columns.tolist())
print(f"\nShape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Clean column names first
df.columns = df.columns.str.strip()

# Replace '?' with NaN across entire dataframe
df = df.replace('?', 0)

# Convert numeric columns to proper numeric type
numeric_cols = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
                'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing values with median or 0
df = df.fillna(0)

target_name = "num"

x = df.drop(columns="num")
y = df[target_name]
x.head()
```

```python
df["oldpeak"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  f1_score, precision_score, recall_score, accuracy_score, confusion_matrix
reg = LogisticRegression(max_iter= 10000)
reg.fit(x_train,y_train)

lr_pred=reg.predict(x_test)

print("accuracy_score:",accuracy_score(y_test,lr_pred))
print("\n F1:\n",f1_score(y_test,lr_pred))
print("\n Precision score is:\n",precision_score(y_test,lr_pred))
print("\n Recall score is:\n",recall_score(y_test,lr_pred))
print( "confusion matrix is:\n",confusion_matrix(y_test,lr_pred))
```

# Model Performance Evaluation: Logistic Regression

Following the data cleaning and preprocessing steps, a Logistic Regression model (max_iter=10000) was trained and tested. Below are the finalized results.

### 1. Key Performance Metrics
| Metric | Score |
| :--- | :--- |
| **Accuracy Score** | **81.36%** |
| **Recall Score** | **88.89%** |
| **F1 Score** | **74.42%** |
| **Precision Score** | **64.00%** |

### 2. Confusion Matrix Analysis
The confusion matrix provides a detailed look at where the model is succeeding and where it is making errors:

| | Predicted: No Disease | Predicted: Disease |
| :--- | :---: | :---: |
| **Actual: No Disease** | 32 (True Negative) | 9 (False Positive) |
| **Actual: Disease** | 2 (False Negative) | 16 (True Positive) |


### 3. Final Observations
* **High Sensitivity:** The model demonstrates a very high **Recall (88.89%)**, which is critical in medical diagnostics. It only missed 2 actual cases of heart disease (False Negatives).
* **Reliability:** With an **Accuracy of 81.36%**, the model is a strong baseline for this dataset.
* **Room for Improvement:** The **Precision (64.00%)** indicates that there are several "False Alarms" (9 cases). Future work could focus on reducing these False Positives to avoid unnecessary follow-up testing for healthy patients.
